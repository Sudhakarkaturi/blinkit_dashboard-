dividing data into catagorical and continuos


```python
def datatypes(df):
  cat_var = []
  cont_var = []
  for col in df:
    if df[col].nunique() <60:
      cat_var.append(col)
    else:
      cont_var.append(col)
  return cat_var, cont_var
```
